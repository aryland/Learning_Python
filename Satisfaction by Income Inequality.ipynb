```python
import pandas as pd

data = pd.read_csv('happiness by income.csv')

print(data)
```

```python
ineq = data['income_inequality']
satisfaction = data['adjusted_satisfaction']

print(satisfaction)
```

```python
import matplotlib.pyplot as plt

plt.scatter(ineq, satisfaction)
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Income Inequality')
plt.ylabel('satisfaction')
plt.scatter(ineq, satisfaction)
```

```python
data.sort_values('income_inequality', inplace=True)

data
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Income Inequality')
plt.ylabel('satisfaction')
plt.scatter(ineq, satisfaction, s = ineq*10, alpha=.15)
```

```python
from sklearn.cluster import KMeans

satisfaction
```

```python
from sklearn.cluster import KMeans
import numpy as np

ineq_satisfaction = np.column_stack((ineq, satisfaction))

km_res = KMeans(n_clusters=5).fit(ineq_satisfaction)

clusters = km_res.cluster_centers_

plt.scatter(ineq, satisfaction, alpha = .25)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)

plt.xlabel('Income Inequality')
plt.ylabel('Satisfaction')
```

```python
most_satisfied = data[data['adjusted_satisfaction'] > 1]

plt.scatter(most_satisfied['income_inequality'], most_satisfied['adjusted_satisfaction'])

plt.text(most_satisfied.iloc[0]['income_inequality'], most_satisfied.iloc[0]['adjusted_satisfaction'], most_satisfied.iloc[0]['country'])
```

```python
from sklearn.cluster import KMeans
import numpy as np

ineq_satisfaction = np.column_stack((ineq, satisfaction))

km_res = KMeans(n_clusters=5).fit(ineq_satisfaction)

clusters = km_res.cluster_centers_

plt.scatter(ineq, satisfaction, alpha = .25)
plt.scatter(clusters[:,0], clusters[:,1], s=1000)

plt.xlabel('Income Inequality')
plt.ylabel('Satisfaction')


most_satisfied = data[data['adjusted_satisfaction'] > 70]

plt.scatter(most_satisfied['income_inequality'], most_satisfied['adjusted_satisfaction'])

plt.text(most_satisfied.iloc[0]['income_inequality'], most_satisfied.iloc[0]['adjusted_satisfaction'], most_satisfied.iloc[0]['country'])
```
